# 1. Data Correction

## 1.1 Drop NA

```python
# tạo bảng df_dropna_all bằng cách loại bỏ tất cả giá trị null trong bảng và in ra info
df_dropna_all = df.dropna(axis=0, how='any')
df_dropna_all
```

```python
# tạo bảng df_dropna_online_order bằng cách loại bỏ giá trị null trong cột online_order và in ra info
df_dropna_online_order = df.dropna(subset=['online_order'])
df_dropna_online_order
```

## 1.2 Fill NA

```python
# fill tất cả NA bằng 0, lưu vào bảng df_fillna_0 và in ra info
df_fillna_0 = df.fillna(value = 0)
df_fillna_0
```

```python
# fill NA bằng ffill, lưu vào bảng df_fillna_ffill và in ra info
df_fillna_ffill = df.fillna(method = 'ffill')
df_fillna_ffill
```

```python
# fill NA bằng bfill, lưu vào bảng df_fillna_bfill và in ra info
df_fillna_bfill = df.fillna(method = 'bfill')
df_fillna_bfill
```

## 1.3. Assess Categorial Data

```python
# load sheet Demographic và hiển thị top head
demographic = pd.read_excel (r'C:\Users\Admin\Downloads\KPMG\KPMG\KPMG.xlsx',sheet_name='CustomerDemographic')
demographic.head()
```

```python
# in ra info bảng 
demographic.info()
```

```python
# describe các giá trị object
demographic.describe(include=['object'])
```

```python
# Remove all NA row
demographic= demographic.dropna(axis=0, how='any')
demographic
```

```python
# In ra các giá trị unique của cột gender
demographic['gender'].unique()
```

```python
# Quy chuẩn cột Gender về 2 nhóm Male và Female, Unisex và in ra giá trị unique mới
demographic = demographic.replace (to_replace = ['F','Femal'], value ='Female')
demographic['gender'].unique()
```

```python
# Tính tuổi của mỗi khách hàng và lưu vào cột "Age" và in ra min age, max age
demographic['Year'] = demographic['DOB'].dt.year
demographic['Age'] = 2022 - demographic['Year']
print(demographic['Age'].min())
print(demographic['Age'].max())
```

```python
# phân nhóm khách hàng dưới 35 tuổi là nhóm "Young", Từ 36 - 55 là "Middle", trên 55 là Older
# lưu vào cột "Age_Group" bằng cách sử dụng hàm cut
bins = [0, 35, 55, 200]
labels = ["Young","Middle","Older"]
demographic['Age_Group'] = pd.cut(demographic['Age'], bins=bins, labels=labels)
demographic.head()
```

# 2. Data Completeness

```python
#join data sale vs demographic và in ra những trường hợp missing trong demographic
table = pd.merge(df_dropna_all, demographic, on = 'customer_id',how = 'left')
table
```

# 3. Data Aggregation

```python
# tính tổng khách hàng bằng hàm nunique
total_customers = table['customer_id'].nunique()
total_customers
```

```python
# Tính số khách khách hàng mỗi tháng
# hint 1: tạo collum year_month bằng hàm dt.strftime
# hint 2: group by year_month và nunique
table['year_month'] = table['transaction_date'].dt.strftime ('%Y-%m')
table
table.groupby('year_month').nunique()['customer_id']
```

```python
# Tính Gross magin của mỗi tháng. Gross magin = list_price - standard_cost
# hint 1: tạo collum Gross_Margin
# hint 2: group by year_month và sử dụng hàm agg
table['Gross magin'] = table['list_price'] - table['standard_cost']
table.groupby('year_month').sum()['Gross magin']
```

```python
# Tính số lượng order và doanh thu group by online_order và order_status 
# hint 1: group by online_order, order_status và sử dụng hàm agg
table.groupby(['online_order', 'order_status'])['online_order'].count()
```

```python
table['revenue'] = table['list_price'] * table['online_order']
table.groupby(['online_order', 'order_status']).sum()['revenue']
```

# 4. Data Visualization

```python
# Sử dụng seaborn và vẽ ra 5 loại chart khác nhau (column, line, box,...) với đầy đủ tên chart, tên các trục và chú thích.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Biểu đồ thể hiện số lượng khách hàng mỗi tháng
g1 = sns.countplot(x = 'Age_Group', data = table, hue='gender')
g1.set_title("Number of customers by age group and gender")
g1.set(xlabel="Age group", ylabel="Number of customers")
plt.show()
```

```python
# Biểu đồ phân bố độ tuổi người sở hữu xe và không sở hữu xe, thể hiện qua giới tính
sns.set_palette('Blues')
g2 = sns.catplot(x='gender', y='Age', data=table, kind="box", hue="owns_car")
g2.fig.suptitle("Age of those owned car vs not")
plt.show()
```

```python
# Biểu đồ phân tán thể hiện doanh thu của các thương hiệu xe đạp trong năm 2017
g3 = sns.catplot(x='brand', y='revenue', data=table, kind='bar')
g3.fig.suptitle("Revenue of bicycle brands in 2017")
g3.set(xlabel="Bicycle brands", ylabel="Revenue in 2017")
plt.xticks(rotation=90)
plt.show()
```

```python
# Biểu đồ thể hiện Gross margin mỗi tháng trong năm 2017
category_order = ['2017-01', '2017-02', '2017-03', '2017-04', '2017-05', '2017-06', '2017-07', '2017-08','2017-09', '2017-10','2017-11','2017-12']
g4 = sns.catplot(x='year_month', y='Gross magin', data=table, kind = 'point', order= category_order)
g4.fig.suptitle("Gross margin in 2017")
g4.set(xlabel="Month in 2017", ylabel="Gross margin per month")
plt.xticks(rotation=90)
plt.show()
```

```python
# Biểu đồ phân tán thể hiện mối tương quan giữa revenue và gross margin của các thương hiệu xe đạp trong năm 2017
g5 = sns.relplot(y="revenue", x="Gross magin", data=table, kind="scatter", col='brand')
g5.fig.suptitle("Revenue vs Gross Margin in 2017", y=1.2)
plt.show()
```
